```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Part A

```python
def gradient_descent(x1, x2, gamma = 1, learning_rate = 0.1, plot = False):

    iterations = 0
    val_arr = [x1**2 + gamma*(x2**2) - x1*x2 - x1 - x2]
    x = 100000
    while True:
        x-=1
        if x==0:
            print("Not Converging")
            break
        dx1 = 2 * x1 - x2 - 1
        dx2 = 2 * gamma * x2 - x1 - 1
        
        x1 = x1 - learning_rate * dx1
        x2 = x2 - learning_rate * dx2
        
        iterations += 1
        try:
            val_arr.append(x1**2 + gamma*x2**2 - x1*x2 - x1 - x2)
        except:
            print("Not Converging")
            val_arr = val_arr[:100]
            break
        if abs(dx1) + abs(dx2) < 0.00000000000001:
            break
    
    if plot:
        plt.plot(val_arr)
        plt.xlabel('Iterations')
        plt.ylabel('Value')
        plt.show()
        
    return x1, x2, iterations
        
```

#### Part B

```python
x1, x2, iterations = gradient_descent(100, 100, learning_rate = 0.1)
print('x1 =', x1, ', x2 =', x2, ', iterations =', iterations)
```

#### Part C

```python
x1, x2, iterations = gradient_descent(100, 100, learning_rate= 0.1, plot=True)
```

#### Part D

```python
learning_rates = [0.025, 0.05, 0.1, 0.25, 0.5,1]
iterations_arr = []
for learning_rate in learning_rates:
    x1, x2, iterations = gradient_descent(100, 100, 1, learning_rate)
    print('x1 =', x1, ', x2 =', x2, ', learning_rate=', learning_rate ,', iterations =', iterations)
    iterations_arr.append(iterations)

plt.plot(learning_rates, iterations_arr)
plt.xticks(learning_rates,rotation = 'vertical')
plt.xlabel('Learning Rate')
plt.ylabel('Iterations')
plt.title('Iterations vs Learning Rate')
```

#### Part E

```python
gammas = [0.25, 0.35, 0.5, 0.7, 0.8, 1]
iterations_arr = []
gn = []
for gamma in gammas:
    x1, x2, iterations = gradient_descent(100, 100, gamma = gamma,learning_rate=0.1)
    print('x1 =', x1, ', x2 =', x2, ', γ =', gamma, ', iterations =',
          iterations if iterations != -1 else 'Not Converging')
    if iterations !=-1:
        iterations_arr.append(iterations)
        gn.append(gamma)

plt.plot(gn, iterations_arr)
plt.xticks(gn, rotation='vertical')
plt.xlabel('γ')
plt.ylabel('Iterations')
plt.title('γ vs Iterations')
```
